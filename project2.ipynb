## Section 4 - Data Science
I think it would be very intersting to find out some things that require a little bit more code to get to. I want to see which federation has the most stadiums. I also want to see which country has the most stadiums. 

```python
df2 = df1.groupby('Confederation').agg('count').reset_index()
LargestConc = df2.nlargest(1,'City')
LargestConc
```

```python
SmallestConc = df2.nsmallest(1,'City')
SmallestConc
```

Here we can see the largest confederation is by far UEFA, or the Union of European Football Associations. The smallest is the OFC, or the Oceania Football Confederation. This makes a lot of sense considering how popular football is in Europe. Soccer isn't known to be too popular in Oceania, so it makes sense how we would only see just a handful of stadiums. Lets do the same with countries, and see what the results are. I predict the largest country will be England.

```python
df2 = df1.groupby('Country').agg('count').reset_index()
LargestCount = df2.nlargest(5,'City')
LargestCount
```

```python
SmallestCount = df2.nsmallest(5,'City')
SmallestCount
```

This is very interesting! It is expected to see England, Spain, Germany, and France in the top 5. However, I did not expect Poland to be #1. They actually have 5 tiers of their professional football system, and even more moving down to the regional level, so it does make sesne that they would have a lot of stadiums. The smallest countries all seem to be tied with 1, and I'm sure this list would continue on. Some countries don't even have an actual professional football proram, so they might just have 1 or 2 stadiums reserved for their national teams.
